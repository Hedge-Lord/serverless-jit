```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV
df = pd.read_csv('../out.csv')

# Convert ns to seconds for readability
df['duration_s'] = df['duration_ns'] / 1e9
df['uptime_s'] = df['uptime_ns'] / 1e9

# Plot uptime vs duration
plt.figure(figsize=(8, 5))
plt.scatter(df['uptime_s'], df['duration_s'])
plt.xlabel('Uptime (s)')
plt.ylabel('Duration (s)')
plt.title('Duration vs Uptime')
plt.show()
```
